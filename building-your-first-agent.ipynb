```python
import os
from getpass import getpass

# Set the API Key
# Replace "OPENAI_API_KEY" with "GROQ_API_KEY" or others if you use a different provider.
os.environ["OPENAI_API_KEY"] = getpass("Enter your OpenAI API Key: ")
```

```python
from crewai import Task, Agent, Crew

content_creator_assistant = Agent(
    role="Content Creator Assistant",
    goal="Come up with video ideas",
    backstory="Someone who has a lot of experience creating video content.",
    llm="gpt-4o",
)

task = Task(
    description="Come up with 5 new video ideas.",
    expected_output="List of ideas.",
    agent=content_creator_assistant
)

crew = Crew(agents=[content_creator_assistant], tasks=[task])

result = crew.kickoff()

print(result.raw)
```

```python
creator_agent = Agent(
    role="YouTube Shorts Micro-History Strategist",
    goal="Plan a 1-week slate of high-retention YouTube Shorts about surprising origins",
    backstory=(
        "You specialize in 30-45s micro-history that hooks fast, pays off with a twist, "
        "You keep ideas filmable by a solo creator at home with minimal props."
    ),
    llm="gpt-4o",
    verbose=True
)

task = Task(
    description=(
        "Create a 1-week video posting plan with 5 video blueprints. "
        "Platform: YouTube Shorts (vertical 9:16, 30-45s). "
        "Niche: Micro-History of Everyday Things (e.g., why pencils are yellow, origins "
        "Primary goals: 1) thumb-stop hook in first 1s, 2) crystal-clear narrative with "
        "3) strong SEO phrasing in title/caption, 4) comment-bait CTA. "
        "Context: solo creator, home-filmable, no special gear. "
    ),
    expected_output=(
        """
        Output a JSON array following the schema below, which contains a
        weekly schedule and 5 video blueprints. Each video blueprint should include:
        {
            "videos": [
                {
                    "title": "<searchable, curiosity-driven title>",
                    "hook_main": "<=12 words, shows payoff fast>",
                    "hook_alt": "<variant hook>",
                    "visuals": ["simple prop or b-roll idea 1", "idea 2"],
                    "tags": ["#microhistory","#everydaythings","#shorts"],
                    "cta": "<question that invites comments>"
                }
            ]
        }
        """
    ),
    agent=creator_agent
)

crew = Crew(agents=[creator_agent], tasks=[task])

result = crew.kickoff()

print(result.raw)
```
